```python

# ======================================================
# üßÆ VARIABLES - Tipus, assignaci√≥ m√∫ltiple i conversions
# ======================================================

print("## VARIABLES ##")

# Tipus b√†sics
nom = "Clara"             # str
edat = 25                 # int
altura = 1.68             # float
estudiant = True          # bool

print("Nom:", nom)
print("Edat:", edat)
print("Altura:", altura)
print("√âs estudiant?", estudiant)

# Assignaci√≥ m√∫ltiple
a, b, c = 1, 2, 3
print("Assignaci√≥ m√∫ltiple: a =", a, ", b =", b, ", c =", c)

# Canvi de tipus (casting)
edat_text = str(edat)      # int ‚Üí str
altura_int = int(altura)   # float ‚Üí int (trunca el decimal)

print("Edat com a text:", edat_text)
print("Altura com a enter:", altura_int)

# Tipus avan√ßats
llista = [1, 2, 3]
diccionari = {"nom": nom, "edat": edat}
conjunt = set([1, 2, 2, 3])  # Elimina duplicats

print("Llista:", llista)
print("Diccionari:", diccionari)
print("Conjunt (sense duplicats):", conjunt)
print("-" * 60)

# ========================================================
# ‚ûï OPERADORS MATEM√ÄTICS - b√†sics, avan√ßats i aplicacions
# ========================================================

print("## OPERADORS MATEM√ÄTICS ##")

x = 12
y = 5

# B√†sics
print(f"Suma: {x} + {y} =", x + y)
print(f"Resta: {x} - {y} =", x - y)
print(f"Multiplicaci√≥: {x} * {y} =", x * y)
print(f"Divisi√≥: {x} / {y} =", x / y)

# Avan√ßats
print(f"Divisi√≥ sencera: {x} // {y} =", x // y)
print(f"Resta (m√≤dul): {x} % {y} =", x % y)
print(f"Pot√®ncia: {x} ** {y} =", x ** y)

# Aplicaci√≥ pr√†ctica: calcular l‚Äô√†rea d‚Äôun cercle
import math
radi = 3
area = math.pi * radi ** 2
print("√Ärea d‚Äôun cercle amb radi 3:", area)
print("-" * 60)

# ========================================================
# üß† OPERADORS L√íGICS - comparaci√≥, combinaci√≥ i aplicacions
# ========================================================

print("## OPERADORS L√íGICS ##")

nota = 8
assistencia = 90

# Comparaci√≥
print("nota > 5:", nota > 5)
print("assistencia >= 80:", assistencia >= 80)

# Combinaci√≥
aprovat = nota >= 5 and assistencia >= 75
excel¬∑lent = nota >= 9 or assistencia == 100

print("Est√† aprovat?", aprovat)
print("√âs excel¬∑lent?", excel¬∑lent)

# Negaci√≥
print("No est√† aprovat?", not aprovat)

# Exemple pr√†ctic
if nota >= 5 and assistencia >= 75:
    print("L‚Äôestudiant aprova.")
else:
    print("L‚Äôestudiant susp√®n.")

# √ös de par√®ntesis per controlar prioritat
print("Exemple amb l√≤gica combinada:", (nota >= 5 and assistencia >= 75) or nota == 10)
print("-" * 60)

# =======================================================
# üé§ ENTRADA I SORTIDA DE DADES - tipus, validaci√≥ i format
# =======================================================

print("## ENTRADA I SORTIDA ##")

# Entrada de dades
nom_usuari = input("Com et dius? ")
print(f"Hola, {nom_usuari}!")

# Entrada amb conversi√≥ a enter
try:
    edat = int(input("Quants anys tens? "))
    print("L‚Äôany que ve tindr√†s", edat + 1, "anys.")
except ValueError:
    print("‚ö†Ô∏è Si us plau, escriu un n√∫mero enter per a l‚Äôedat.")

# Entrada amb conversi√≥ a float
try:
    pes = float(input("Quin √©s el teu pes en kg? "))
    print("Pes registrat:", pes, "kg")
except ValueError:
    print("‚ö†Ô∏è Si us plau, escriu un n√∫mero decimal per al pes.")

# Sortida formatada
altura = 1.72
print(f"{nom_usuari}, mesures registrades: {altura} m d'altura i {pes:.1f} kg de pes.")
print("-" * 60)
```
